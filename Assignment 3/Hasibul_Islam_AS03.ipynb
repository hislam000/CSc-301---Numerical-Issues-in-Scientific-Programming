```python
import math;

# Calculate the Taylor Polynomials: 0 for functions at the values of x
from math import factorial, sin, cos, exp, log
import matplotlib.pyplot as plt
from decimal import *

#value of x
#number of n
def taylor_exp(x, n):
    j=0
    total= 0.0
    while j <= n:
        total = Decimal(total) + Decimal((x**j)/(factorial(j)))
        j += 1
    return total
```
